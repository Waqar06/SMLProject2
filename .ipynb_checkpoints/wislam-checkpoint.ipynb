```python
def train_logistic_regression(X, y, **args):
    """
    Train a logistic regression model on dataset (X, y) and return trained model.
    X: matrix of real values, size n x d
    y: vector of string labels, size n
    args: optional arguments e.g., for hyper-parameters
    """
    # Training  Logistic Regression Model 
    logistic_regression = LogisticRegression(multi_class='multinomial',C=0.01,penalty='l2',solver='lbfgs')
    model = logistic_regression.fit(X, y)
    return model
```

```python
def evaluate_logistic_regression_accuracy(Xt, yt, model):
    """
    Apply logistic regression prediction on dataset Xt and evaluate accuracy against yt,
    returing the accuracy results as a scalar.
    Xt: matrix of real values, size m x d
    yt: vector of string labels, size m
    """
    # Evaluating the accuracy of model against a test set
    # Using "accuracy_score" from sklearn. Accuracy classification score. In multilabel classification, this 
    # function computes subset accuracy:      
    # the set of labels predicted for a sample must exactly match the corresponding set of labels in y_true()
    yLabel_prediction = model.predict(Xt)
    score = accuracy_score(yt, yLabel_prediction)
    return score

  
```

## 2. Active learning framework with Random selection

```python
def random_select(X, model, **args):
    """
    Given an unlabelled dataset X, a matrix of n x d, and a model (not used)
    returns a vector of scores of length n. Each entry reflects the priority 
    of the corresponding instance. Higher means better.
    """
    # Creating a score array using random numbers of length of X 
    scores = np.random.randint(1,len(X),len(X))
    return scores
    
```

```python
def pool_based_active_learning(X_pool, y_pool, seed_ids,
                               train_func, select_func,
                               max_size, batch_size, **args):
    """
    Perform an active learning simulation, which starts by training on a seed set,
    then iteratively applies the selection function to rank instances in the pool,
    selects the top few instances which are included into the training set and the
    process repeats. 
        X_pool: matrix of n x d
        y_pool: vector of string labels, size n
        seed_ids: initial labelled set set, as a list of indices [0..n-1] into pool
        train_func: function which given (X, y, optional args) returns a trained model
        select_func: function which given (X, optional args) returns a sequence of scores
        max_size: stopping condition for active learning, when labelled data reaches given size
        batch_size: number of instances to be labelled in each iteration
        args: optional arguments passed to training and selection function
    returns the sequence of trained models 
    """
    if (batch_size <= 0):
        raise ValueError("Batch Size should be positive real number")
    if (max_size <= 0):
        raise ValueError("Max Size should be positive real number")


    trained_models =[]       # list to save the sequence of trained models
    seed = copy(seed_ids)    # avoid updating original seed_set
    U =  np.delete(X_pool,seed, axis=0)   # pool of unlabelled instances,
    L = [X_pool[seed],y_pool[seed]]       # set of initial labelled instances L = (x,y)  
    b =  batch

    while(len(seed) < max_size):          # max_size: max size of instances in set of initial labelled instances  
        Theta_model=train_func(L[0],L[1]) # train Function 
   
        r = select_func(U,Theta_model)    # score Function
        max_indices = np.argpartition(r,-b)[-b:]   #return Top (Batch=60) instances indices according to score
        for j in max_indices:
            L[0] = np.append(L[0],[X_pool[j]], axis=0)    # update set of initial labelled instances  
            L[1]= np.append(L[1],y_pool[j])               # update set of initial labelled instances
            seed.append(j)
            U = np.delete(X_pool,seed,axis=0)             # upadte pool of unlabelled instances -U
        trained_models.append(Theta_model)                
    
    return trained_models                                 # returns a list of sequence of trained models

```

## 3. Uncertainty sampling

```python
def logistic_regression_entropy_select(X, model, **args):
    """
    Given an unlabelled dataset X, a matrix of n x d, and a discriminative model 
    P(y|x), returns a vector of n entropy values.
    """
    # calculating entropy values for given model.   
    prob = model.predict_proba(X)
    ent  = entropy(prob,axis=1)
    return ent
```

## 4. Query by committee

```python
# don't forget to provide function descriptive comments, like those provided in templates above

def query_by_committee_vote_entropy(X, model, **args):
    """
    Perform QUERY BY COMMITTEE, using vote entropy which is a heuristics for measuring disagreement.
    This committee-based generalization of uncertainty measures is also known as hard vote.
    Given an unlabelled dataset X, a matrix of n x d, and a model
    returns a vector of scores of length n with high entropy of "most informative" samples
    This function takes following inputs.   
        X: matrix of n x d
        model: tuple of model which contains ensemble models and base model
    returns the length n vector of entropy values
    """
    ensemble_model_list = model[0]        # get ensemble model list from tuple
    yLabel_classes_len = len(ensemble_model_list[0].classes_)  # get unique number of classes from y_lable list
    committee_size= len(model[0])         # get committee_size 'C' of ensemble
    predictions_yLabel_list = []
    for model in ensemble_model_list:
        predictions_yLabel = model.predict(X)    # getting predicted Label using each committee member (Model-ensemble)
        predictions_yLabel_list.append(np.eye(yLabel_classes_len)[predictions_yLabel])  # create a matrix where unique no.classes are column and each X is row. 
                                                                                        # All entries in a row are zero, and store 1 where to label column which 
                                                                                        # which is assigned by model    
    # Both matrix are added, to get sum of votes for each class assigned by ensemble.Dividing sum by committee size to get average. 
    votes_classes = np.apply_along_axis(np.sum, 0, np.stack(predictions_yLabel_list)) / committee_size 
    scores_hard_vote = np.apply_along_axis(entropy, 1, votes_classes)   # Calculate entropy from vote total count       
    
    return scores_hard_vote
    

def query_by_committee_soft_vote_entropy(X, model, **args):
 
    """
    Perform QUERY BY COMMITTEE, using vote entropy which is a heuristics for measuring disagreement.
    This committee-based generalization of uncertainty measures which accounts for 
    each committee member’s conﬁdence: is also known as soft vote

    Given an unlabelled dataset X, a matrix of n x d, and a model
    returns a vector of scores of length n with high entropy of "most informative" samples
    This function takes following inputs.   
        X: matrix of n x d
        model: tuple of model which contains ensemble models
    returns the length n vector of entropy values
    """
    ensemble_model_list = model[0]   # get ensemble model list from tuple
    committee_size= len(model[0])   # get committee_size 'C' of ensemble
    
    predictions_yLabel_list = []
    for model in ensemble_model_list:
        predictions_yLabel = model.predict_proba(X) # getting predicted probability of each class using each committee member (Model-ensemble)
        predictions_yLabel_list.append(predictions_yLabel)  # storing the matrix in a list for each ensemble model. 
        

    # # Both matrix are added, to get sum of probabilities for each class assigned by ensemble.Dividing sum by committee size to get average.
    votes_classes = np.apply_along_axis(np.sum, 0, np.stack(predictions_yLabel_list)) / committee_size
    scores_soft_vote = np.apply_along_axis(entropy, 1, votes_classes)    # Calculate entropy of average probablity 

    return scores_soft_vote

def query_by_committee_KL(X, model, **args):
 
    """
    Perform QUERY BY COMMITTEE, using vote entropy which is a heuristics for measuring disagreement.
    Kullback-Leibler (KL) divergence is an information-theoretic measure of the difference between two probability                                  distributions. In this case, we want to quantify disagreement as the average divergence of each committee member θ’s prediction from that       of the consensusC: 
    
    Given an unlabelled dataset X, a matrix of n x d, and a model
    returns a vector of scores of length n with high entropy of "most informative" samples
    This function takes following inputs.   
        X: matrix of n x d
        model: tuple of model which contains ensemble models
    returns the length n vector of entropy values
    """
    ensemble_model_list = model[0]    # get ensemble model list from tuple
    committee_size= len(model[0])     # get committee_size 'C' of ensemble
    
    predictions_yLabel_list = []
    for model in ensemble_model_list:
        predictions_yLabel = model.predict_proba(X)  # getting predicted probability of each class using each committee member (Model-ensemble)
        predictions_yLabel_list.append(predictions_yLabel)  # storing in a list for both ensemble model. 
        

    consensus_mean_matrix = np.mean(np.stack(predictions_yLabel_list), axis=0) #calculating mean from using both matrix in from the list.(matrix of probablities for each class)
    divergence_list = []
    for predictions_list in predictions_yLabel_list:
        ent = entropy(consensus_mean_matrix.T, predictions_list.T)  #calculating entropy of divergence matrix- taking transpose so sequance is same for entropy calculation 
        divergence_list.append(ent)   # storing entropy for each ensemble model 
    scores_kl_vote = np.apply_along_axis(np.mean, 0, np.stack(divergence_list)) # taking mean for entropy values.  
    
    return scores_kl_vote

```

```python
def train_committee(X, y, **args):
    
    """
    Train Committee method trains 1 Ensemble model which contains two model and one base logistic regression 
    Base Logistic regression model on dataset (X, y) and Ensemble models are trained on dataset (X, y) by using Bagging technique. 

    Given an unlabelled dataset X, a matrix of n x d, and y of length n.
    returns a tuple which contains Ensemble of 2 memebers and base logistic regression
    This function takes following inputs.   
        X: matrix of n x d
        y: of length n
      
    returns a tuple of model which contains ensemble models and base model
    """
    # Training Base Logistic Regression model 
    model_base = train_logistic_regression(X,y)

    max_sample = len(X)          # Length of samples to be used as size of sub-samples
    encoded_list_y = encode_Y(y) # Encoding y_labels, to get label ids instead of labels

    # Bagging Ensemblers
    # Committee member 1                             
    # Paramter Settings   
    # 1- bootstrap=True(makes sample by replacement for bagging), 
    # 2- max_samples=max_sample(which makes sample of length of X) 
    # 3- C=0.01 To differentiate from model committee_member_2
    # 4- random_state=42 To differentiate from committee_member_2
    
    bag_ensemble_c1 = BaggingClassifier(LogisticRegression(random_state=42,multi_class='multinomial',C=0.01,penalty='l2',solver='lbfgs'),
                                        n_estimators=1,max_samples=max_sample,oob_score=False, bootstrap=True, random_state=42)
    
    committee_member_1 = bag_ensemble_c1.fit(X, encoded_list_y)

    # Committee member 2
    # Paramter Settings   
    # 1- bootstrap=True(makes sample by replacement for bagging), 
    # 2- max_samples=max_sample(which makes sample of length of X) 
    # 3- C=0.01 To differentiate from committee_member_1
    # 4- random_state=45 To differentiate from committee_member_1

    bag_ensemble_2 = BaggingClassifier(LogisticRegression(random_state=42,multi_class='multinomial',C=0.01,penalty='l2',solver='lbfgs'),
                                       n_estimators=1,max_samples=max_sample, bootstrap=True,oob_score=False, random_state=45)#45
    
    committee_member_2 = bag_ensemble_2.fit(X, encoded_list_y)
    
    learner_tuple = ([committee_member_1,committee_member_2],model_base)   
    return learner_tuple

    
```
